```python
"""
8. Ev Otomasyon Sistemi
Problem Tanımı: Akıllı ev cihazlarını (ışık, termostat, güvenlik kamerası) 
yöneten bir sistem oluşturun. Her cihaz açılıp kapatılabilir, durum bilgisi 
verebilir ve özel işlevlerini yerine getirebilir.
Tüm cihazlar bir ev sistemi tarafından yönetilmelidir.
"""
class Cihaz:
    def __init__(self, konum):
        self.konum = konum

    def ac(self):
        pass

    def kapat(self):
        pass

    def durum(self):
        pass

class Isik(Cihaz):
    def __init__(self, konum, acik=False):
        super().__init__(konum)
        self.acik = acik

    def ac(self):
        self.acik = True

    def kapat(self):
        self.acik = False

    def durum(self):
        print(f"{self.konum} ışığı {'açık' if self.acik else 'kapalı'}")

class Termostat(Cihaz):
    def __init__(self, konum, sicaklik=20):
        super().__init__(konum)
        self.sicaklik = sicaklik

    def ayarla(self, sicaklik):
        self.sicaklik = sicaklik

    def durum(self):
        print(f"{self.konum} sıcaklığı {self.sicaklik}°C")


salon_isik = Isik("Salon", True)
oturma_term = Termostat("Oturma", 22)
salon_isik.durum()
oturma_term.durum()
```

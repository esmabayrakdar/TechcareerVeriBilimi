```python
"""
2. Hesap Makinesi Sınıfı
Problem Tanımı: Temel matematik işlemlerini gerçekleştiren 
bir hesap makinesi sınıfı oluşturun. 
Sınıf toplama, çıkarma, çarpma, bölme ve bellek işlemlerini desteklemelidir.
"""

class HesapMakinesi:
    def __init__(self):
        self.bellek = 0

    def topla(self, a, b):
        sonuc = a + b
        self.bellek = sonuc
        return sonuc

    def cikar(self, a, b):
        sonuc = a - b
        self.bellek = sonuc
        return sonuc

    def carp(self, a, b):
        sonuc = a * b
        self.bellek = sonuc
        return sonuc

    def bol(self, a, b):
        if b == 0:
            raise ValueError("Hata: Sıfıra bölme mümkün değil!")
        sonuc = a / b
        self.bellek = sonuc
        return sonuc

    def bellek_goster(self):
        return self.bellek

    def bellek_sifirla(self):
        self.bellek = 0
        print("Bellek sıfırlandı.")


# Ana program
hesap = HesapMakinesi()

while True:
    print("\n--- Hesap Makinesi ---")
    print("1. Toplama")
    print("2. Çıkarma")
    print("3. Çarpma")
    print("4. Bölme")
    print("5. Belleği Göster")
    print("6. Belleği Sıfırla")
    print("7. Çıkış")

    secim = input("İşlem seçiniz (1-7): ")

    if secim == "7":
        print("Program sonlandırıldı.")
        break

    if secim in ["1", "2", "3", "4"]:
        try:
            a = float(input("Birinci sayıyı girin: "))
            b = float(input("İkinci sayıyı girin: "))
        except ValueError:
            print("Lütfen geçerli bir sayı girin!")
            continue

        if secim == "1":
            print("Sonuç:", hesap.topla(a, b))
        elif secim == "2":
            print("Sonuç:", hesap.cikar(a, b))
        elif secim == "3":
            print("Sonuç:", hesap.carp(a, b))
        elif secim == "4":
            try:
                print("Sonuç:", hesap.bol(a, b))
            except ValueError as e:
                print(e)

    elif secim == "5":
        print("Bellekteki değer:", hesap.bellek_goster())

    elif secim == "6":
        hesap.bellek_sifirla()

    else:
        print("Geçersiz seçim! Lütfen 1-7 arasında bir değer girin.")
```

```python
"""
9. E-ticaret Sepet Sistemi
Problem Tanımı: E-ticaret sitesi için ürün, sepet ve sipariş yönetimi oluşturun.
Ürünler katalogda bulunabilir, sepete eklenebilir ve sipariş haline getirilebilir. 
Sepet toplamı, indirim uygulaması ve ödeme durumu takibi olmalıdır.
"""
class Urun:
    def __init__(self, isim, fiyat, stok):
        self.isim = isim
        self.fiyat = fiyat
        self.stok = stok

class Sepet:
    def __init__(self):
        self.urunler = []

    def ekle(self, urun, adet):
        if adet > urun.stok:
            print(f"{urun.isim} stokta yetersiz!")
        else:
            self.urunler.append((urun, adet))
            urun.stok -= adet

    def toplam_tutar(self):
        return sum(urun.fiyat*adet for urun, adet in self.urunler)

    def indirim_uygula(self, yuzde):
        tutar = self.toplam_tutar()
        indirim = tutar * yuzde/100
        return tutar, indirim, tutar-indirim


urun1 = Urun("Laptop", 5000, 10)
sepet = Sepet()
sepet.ekle(urun1, 2)
toplam, indirim, odenecek = sepet.indirim_uygula(10)
print(f"Sepet Toplamı: {toplam} TL, İndirim: {indirim} TL, Ödenecek: {odenecek} TL")
```

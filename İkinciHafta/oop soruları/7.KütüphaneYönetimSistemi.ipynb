```python
"""
7. Kütüphane Yönetim Sistemi
Problem Tanımı: Kitap, üye ve kütüphane yönetimi için bir sistem oluşturun. 
Kitaplar ödünç verilebilir, üyeler kitap alabilir ve 
kütüphane işlemlerini takip edebilir. Her kitap ve üye için durum takibi yapılmalıdır.
"""
class Kitap:
    def __init__(self, isim, yazar):
        self.isim = isim
        self.yazar = yazar
        self.odunc_verildi = False
        self.kime_verildi = None

class Uye:
    def __init__(self, isim):
        self.isim = isim
        self.odunc_kitaplar = []

class Kutuphane:
    def __init__(self):
        self.kitaplar = []

    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)

    def odunc_ver(self, kitap, uye):
        if kitap.odunc_verildi:
            print(f"{kitap.isim} zaten ödünç verilmiş.")
        elif len(uye.odunc_kitaplar) >= 3:
            print(f"{uye.isim} maksimum kitap sayısına ulaştı.")
        else:
            kitap.odunc_verildi = True
            kitap.kime_verildi = uye
            uye.odunc_kitaplar.append(kitap)
            print(f"{kitap.isim} kitabı {uye.isim}'e verildi. Kütüphanede kalan: {len(self.kitaplar)-len([k for k in self.kitaplar if k.odunc_verildi])} kitap")


kütüphane = Kutuphane()
kitap1 = Kitap("Python", "Ali")
kitap2 = Kitap("C++", "Veli")
uye1 = Uye("Ahmet")
kütüphane.kitap_ekle(kitap1)
kütüphane.kitap_ekle(kitap2)
kütüphane.odunc_ver(kitap1, uye1)
```

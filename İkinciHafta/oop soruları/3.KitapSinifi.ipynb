```python
"""
3. Kitap Sınıfı
Problem Tanımı: Bir kitabın başlığı, yazarı, sayfa sayısı ve okunma durumu gibi
özelliklerini yöneten bir sınıf oluşturun. 
Sınıf kitap okuma, sayfa atlama ve kitap bilgilerini gösterme metodlarına sahip olmalıdır.
"""
class Kitap:
    def __init__(self, baslik, yazar, sayfa_sayisi):
        self.baslik = baslik
        self.yazar = yazar
        self.sayfa_sayisi = sayfa_sayisi
        self.__okunan_sayfa = 0

    def sayfa_atla(self, sayfa_sayisi):
        if sayfa_sayisi < 0:
            print("Negatif sayfa atlanmaz!")
            return
        self.__okunan_sayfa += sayfa_sayisi
        if self.__okunan_sayfa >= self.sayfa_sayisi:
            self.__okunan_sayfa = self.sayfa_sayisi
            print(f"Kitap tamamlandı! Toplam sayfa: {self.sayfa_sayisi}")
        else:
            print(f"{sayfa_sayisi} sayfa atlandı. Şu an {self.__okunan_sayfa}. sayfadasınız.")

    def okundu_mu(self):
        return self.__okunan_sayfa == self.sayfa_sayisi
        
    def okunan_sayfa(self):
        return self.__okunan_sayfa
        
    def bilgileri_goster(self):
        print(f"Başlık: {self.baslik}")
        print(f"Yazar: {self.yazar}")
        print(f"Sayfa sayısı: {self.sayfa_sayisi}")
        print(f"Okunan sayfa: {self.__okunan_sayfa}")
        print(f"Kitap okundu mu?: {'Evet' if self.okundu_mu() else 'Hayır'}")


# Örnek kullanım
kitap = Kitap("Suç ve Ceza", "Fyodor Dostoyevski", 500)
kitap.bilgileri_goster()
kitap.sayfa_atla(50)
kitap.sayfa_atla(100)
kitap.sayfa_atla(200)
kitap.bilgileri_goster()
```

```python
"""
1. Öğrenci Sınıfı
Problem Tanımı: Bir öğrencinin adı, öğrenci numarası, notları ve not ortalaması gibi
temel bilgilerini yöneten bir sınıf oluşturun. 
Sınıf, yeni not ekleme, ortalama hesaplama ve bilgileri gösterme metodlarına sahip olmalıdır.
"""

class Ogrenci:
    def __init__(self, ad, ogrNo):
        self.ad = ad
        self.ogrNo = ogrNo
        self.notlar = []

    def notEkle(self,yeni_not):
        if 0<= yeni_not <= 100:
            self.notlar.append(yeni_not)
            print(f"{yeni_not} notu eklendi")
        else:
            print("Hatalı not girdiniz.")

    def ortalamaHesapla(self):
        if len(self.notlar) == 0:
            return 0
        return sum(self.notlar) / len(self.notlar)

    def bilgiler(self):
        print(f"Öğrenci Adı: {self.ad}")
        print(f"Öğrenci Numarası: {self.ogrNo}")
        print(f"Notlar: {self.notlar}")
        print(f"Ortalama: {self.ortalamaHesapla():}")

ogr1 = Ogrenci("Esma Bayrakdar", "20250123")

ogr1.notEkle(90)
ogr1.notEkle(85)
ogr1.notEkle(110)   # Hatalı not örneği
ogr1.notEkle(75)

ogr1.bilgiler()
```

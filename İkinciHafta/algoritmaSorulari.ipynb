```python
# 1. Listedeki en büyük sayıyı bulma
numbers = [20, 30, 35, 45, 60]

def en_buyuk_sayi(numbers):
    enbuyuk = numbers[0]
    for sayi in numbers:
        if sayi > enbuyuk:
            enbuyuk = sayi
    return enbuyuk

print("En büyük sayı:", en_buyuk_sayi(numbers))
 
```

```python
# 3.Faktöriyel hesaplama
n = int(input('negatif olmayan tamsayı giriniz: '))
def faktoriyel(n):
    if n == 0:
        return 1 # 0'ın faktoriyeli 1
    sonuc = 1
    for i in range(2,n+1):
        sonuc*=i
    return sonuc
print('Faktoriyel sonucu:', faktoriyel(n))
```

```python
# 7.İki sayının en büyük ortak böleni(EBOB)
a = int(input('sayi1: '))
b = int(input('sayi2: '))

def ebob(a, b):
    while b:           #öklid algoritması
        a, b=b, a%b
    return a     

print('EBOB: ', ebob(a, b))
    
```

```python
# 8.Anagram kontrolü
def anagramMı(p1,p2):
    p1 = p1.lower().replace(" ","")
    p2 = p2.lower().replace(" ","")
    return sorted(p1) == sorted(p2)

p1 = input('string1: ')
p2 = input('string2: ')

if anagramMı(p1,p2):
    print('bu kelimeler anagram')
else:
    print('bu kelimeler anagram değil')

```

```python
# 9.Listedeki eksik sayıyı bulma
numbers = [1,2,3,4,6,7]
def EksikSayiBul(numbers):
    n = len(numbers)+1

    beklenen_toplam = n*(n+1)//2
    
    gercek_toplam = sum(numbers)
    return beklenen_toplam - gercek_toplam

print('Eksik sayı:',EksikSayiBul(numbers))
```

```python
# 10. Listeyi belirli bir değere göre döndürme
# verilen bir listeyi, k adımı kadar sağa döndüren bir fonksiyon yazın.
liste = [1,2,3,4,5]
k = int(input('ne kadar sağa döndürülsün: '))
def sağaDöndür(liste,k):
    n=len(liste)
    k = k % n
    if k == 0:
        return liste
    return liste[-k:] + liste [:-k]

print(sağaDöndür(liste,k))
```

```python
# 11.iki sıralı listenin medyanını bulma
liste1 = [1,3,4]
liste2 = [2,5,6]

def medyanBul(liste1, liste2):
    birlestir = sorted(liste1 + liste2)
    n = len(birlestir)
    if n%2==1:
        return birlestir[n//2]
    else:
        orta1 = birlestir[n//2 -1]
        orta2 = birlestir[n//2]
        return (orta1+orta2)/2

print("Medyan: ", medyanBul(liste1,liste2))
```

```python
# 12.En uzun Palindromik alt string
def longestPalindrome(s):
    if not s:
        return ""
    
    start = 0
    max_len = 1
    
    def expand_around_center(left, right):
        
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        return right - left - 1  
    
    for i in range(len(s)):
        
        len1 = expand_around_center(i, i)
       
        len2 = expand_around_center(i, i + 1)
        
        
        current_max = max(len1, len2)
        if current_max > max_len:
            max_len = current_max
            start = i - (current_max - 1) // 2
    
    return s[start:start + max_len]


print(longestPalindrome("babad"))
```

```python
# 13.Kelime merdiveni
from collections import deque

def wordLadder(beginWord, endWord, wordList):
    wordSet = set(wordList)
    if endWord not in wordSet:
        return 0

    queue = deque([(beginWord, 1)])

    while queue:
        word, length = queue.popleft()

        if word == endWord:
            return length

      
        for i in range(len(word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                newWord = word[:i] + c + word[i+1:]

                if newWord in wordSet:
                    wordSet.remove(newWord)
                    queue.append((newWord, length + 1))

    return 0 

print(wordLadder("talk", "tail", ["tall", "tail", "balk", "tale"]))
```

```python
# 14.Maksimum alt dizi toplamı
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

def maxSubArray(nums):
    current_sum = max_sum = nums[0]
    for num in nums[1:]:
        current_sum = max(num, current_sum+num)
        max_sum= max(max_sum,current_sum)
    return max_sum
print(maxSubArray(nums))
```

```python
# 15.Birleştirme aralıkları
def araliklari_birlestir(araliklar):
    if not araliklar:
        return []

    araliklar.sort(key=lambda x: x[0])

    birlestirilmis = [araliklar[0]]

    for mevcut_aralik in araliklar[1:]:
        son_aralik = birlestirilmis[-1]

       
        if mevcut_aralik[0] <= son_aralik[1]:
        
            son_aralik[1] = max(son_aralik[1], mevcut_aralik[1])
        else:

            birlestirilmis.append(mevcut_aralik)

    return birlestirilmis

araliklar = [[1,3],[2,6],[8,10],[15,18]]
print(araliklari_birlestir(araliklar))
```
